```python
import torch

a = torch.arange(12).reshape(1,3,4)
b = torch.arange(12,16).reshape(1,1,4)
print(a,b)
```

```python
idx_1d = torch.tensor([1], dtype=torch.long)
idx_3d = idx_1d.view(1,1,1).expand(1,1,4)
print(idx_3d)
```

```python
a.gather(1, idx_3d)
```

```python
a.scatter(1 ,idx_3d, b)
```

```python
print(a)
a[:, [1], :] = b
print(a)
```
