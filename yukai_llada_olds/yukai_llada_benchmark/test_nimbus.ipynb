```python

from transformers import AutoModel, AutoTokenizer

device = 'cuda:0'
name_model = 'distilbert/distilbert-base-uncased'
```

```python

tokenizer = AutoTokenizer.from_pretrained(name_model)
model = AutoModel.from_pretrained(name_model).to(device)
```

```python
import torch
seq = tokenizer(['hello', 'hi', 'father'])['input_ids']
hello, hi, father = torch.tensor(seq[0][1], dtype=torch.long, device=device), torch.tensor(seq[1][1], dtype=torch.long, device=device), torch.tensor(seq[2][1], dtype=torch.long, device=device)
```

```python
# model.embeddings.word_embeddings(query_seq)
e_hello = model.embeddings.word_embeddings(hello).unsqueeze(0)
e_hi = model.embeddings.word_embeddings(hi).unsqueeze(0)
e_father = model.embeddings.word_embeddings(father).unsqueeze(0)

from torch.nn.functional import cosine_similarity

print(cosine_similarity(e_hello, e_hi))
print(cosine_similarity(e_hello, e_father))
```
