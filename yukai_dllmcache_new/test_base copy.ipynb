```python
import inspect
from abc import ABC

class NONO(ABC):
    pass

class Inspector(NONO):
    def check(self):
        frame = inspect.currentframe()

        while type(frame.f_locals.get('self')).__bases__[0] == Inspector: 
            frame = frame.f_back
        # end

        return frame.f_locals.get('self').name
    # end
# end

class A(Inspector):
    def hello(self):
        return self.check()
    # end

    def hello2(self):
        return self.hello()
    # end
# end

class Client:
    def main(self):
        a = A()
        self.name = 'hello'
        print(a.hello2())
    # end
# end

Client().main()
```
