```python
import os
filenames_step = [filename for filename in os.listdir('.') if filename[0] != '.'and 'semi' in filename]
corpus_filename_num = [(filename_step, int(filename_step.split('budget_')[-1].split('_')[0])) for filename_step in filenames_step]
corpus_filename_num_sorted = sorted(corpus_filename_num, key=lambda corpus: corpus[1])

for filename_step, _ in corpus_filename_num_sorted:
    with open(filename_step, 'r') as file:
        lines = [line.lstrip().rstrip() for line in file.readlines()]
    # end

    confs = []

    for line in lines:
        str_conf = line.split(',')[0].split('(')[-1]
        confs.append(float(str_conf))
    # end

    print(f'{filename_step}: {sum(confs)/len(confs)}')
# end
```
