```python

class InspectorPlugin:

    def __bool__(self):
        name_klass = self.__class__ # <class '__main__.A_Enabled'>
        str_enabled = str(name_klass).split('.')[-1].split("'")[0].split('_')[-1].lower()

        return str_enabled == 'enabled'
    # end
# end

class A_Disabled(InspectorPlugin):
    pass
# end

class A_Enabled(InspectorPlugin):
    pass
# end
```

```python
p = A_Disabled()
if p:
    print('enabled')
else:
    print('disabled')
# end
```
