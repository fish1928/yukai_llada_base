```python
import os
import json
import torch



class RefreshIdxHelper:

    TYPE_HIDDEN = {
        'k':'_k_previous',
        'v':'_v_previous'
    }

    def __init__(self, dict_filename_to_list_idx_sorted, type_hidden):
        self.dict_filename_to_list_idx_sorted = dict_filename_to_list_idx_sorted
        self.type_hidden=type_hidden
    # end

    def get_refresh_idx(self, x, id_sample, id_block, budget, return_sorted=True):
        filename = f'batch_{id_sample}{self.type_hidden}.pt'
        list_idx_sorted = self.dict_filename_to_list_idx_sorted[filename]
        result = torch.tensor(list_idx_sorted[id_block]['idx'][:budget], dtype=torch.long, device=x.device)
        return torch.sort(result)[0] if return_sorted else result
    # end
# end

class RefreshIdxHelper_Mocked(RefreshIdxHelper):
    def get_refresh_idx(cls, x, len_prompt=0, type_refresh=None):
        if type_refresh == 'previous_all':
            return torch.arange(len_prompt, dtype=torch.long, device=x.device)
        # end

        return torch.tensor([0,1], dtype=torch.long, device=x.device)
    # end
# end

```

```python
with open('all_in_one_sim_report.json', 'r') as file:
    dict_filename_to_list_idx_sorted = json.load(file)
# end

helper_idx = RefreshIdxHelper(dict_filename_to_list_idx_sorted, 'k')
helper_idx.get_refresh_idx(torch.tensor(1.0, dtype=torch.bfloat16), RefreshIdxHelper.TYPE_HIDDEN['k'], 0, 0, 3, return_sorted=True)
```
